*Data Cleaning*

```python
# To check the missing values
df.isnull().sum()
```

```python
# Replace the NULL Values with Median of their Category
# i.e If the Footware Category Item has missing value it will filled by meedian of Footware category.
# This helps to alloacte the specific NULL values according to the Category. Not with the entire Category Value Median**
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Transform columns into snake case.
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns ={'purchase_amount_(usd)':'purchase_amount'})
```

**Feature Engineering**

```python
# Create Column age_group based on age
label = ['Young_Adult','Adult','Middle_Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels= label)
```

```python
# Create Column for purchase_frequency_days
frequency_mapping ={
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Anually' : 365,
    'Quarterly' : 90, 
    'Bi-weekly' : 14, 
    'Monthly' : 30,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
